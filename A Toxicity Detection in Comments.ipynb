```python
import re
import pandas as pd
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
#import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
from nltk.tokenize import word_tokenize
```

```python
dataset = pd.read_csv('youtoxic_english.csv')
dataset
```

```python

# Sélection de la deuxième colonne
col2 = dataset.iloc[:, 2]

# Affichage du tableau
print(col2)
```

```python


# Fonction de nettoyage de texte
def clean_text(text):
    # Convertir le texte en minuscules
    text = text.lower()
    
    # Retirer la ponctuation
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    
    # Retirer les chiffres
    text = re.sub('\d+', '', text)
    
    # Retirer les stopwords
    stopwords_list = stopwords.words('english')
    tokens = nltk.word_tokenize(text)
    tokens = [word for word in tokens if word not in stopwords_list]
    text = ' '.join(tokens)
    
    # Stemming des mots
    stemmer = PorterStemmer()
    tokens = nltk.word_tokenize(text)
    tokens = [stemmer.stem(word) for word in tokens]
    text = ' '.join(tokens)
    
    return text

# Appliquer la fonction de nettoyage à la deuxième colonne
col2_clean = col2.apply(lambda x: clean_text(x))

# Afficher le résultat
print(col2_clean)
```

```python
from collections import defaultdict
import math

# Exemple d'ensemble de commentaires


# Initialisation du dictionnaire de fréquences
frequences = defaultdict(int)

# Prétraitement des commentaires et calcul des fréquences
for commentaire in col2_clean:
    # Prétraitement : suppression de la ponctuation, mise en minuscules, etc.

    # Diviser le commentaire en mots individuels
    mots = commentaire.split()

    # Calcul des fréquences des mots
    for mot in mots:
        frequences[mot] += 1

# Affichage des fréquences des mots
# for mot, frequence in frequences.items():
#     print(mot, ":", frequence)
# Calcul de l'IDF pour chaque terme
N = len(col2_clean)
idf = {}
for mot, frequence in frequences.items():
    idf[mot] = math.log(N / (frequence + 1))

# Calcul de la pondération TF-IDF pour chaque terme dans chaque commentaire
for commentaire in col2_clean:
    tfidf = defaultdict(float)
    mots = commentaire.split()
    for mot in mots:
        tf = frequences[mot] / len(mots)
        tfidf[mot] = tf * idf[mot]

    # Affichage des pondérations TF-IDF pour le commentaire
    print(commentaire)
    for mot, poids in tfidf.items():
        print(mot, ":", poids)
    print()
```

# clustring avec l'algorithme k-means

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import pandas as pd

# Calcul des vecteurs TF-IDF
vectorizer = TfidfVectorizer()
tfidf_vectors = vectorizer.fit_transform(col2_clean)

# Création d'un modèle de clustering K-means avec 2 clusters
kmeans = KMeans(n_clusters=2, random_state=1000)

# Entraînement du modèle sur les vecteurs TF-IDF
kmeans.fit(tfidf_vectors)

# Obtention des étiquettes de cluster pour chaque commentaire
labels = kmeans.labels_

# Création d'un nouveau dataframe contenant les commentaires et leur étiquette de cluster
clustered_data = pd.DataFrame({'Commentaire': col2_clean, 'Cluster': labels})

# Affichage des commentaires et leur étiquette de cluster
for _, row in clustered_data.iterrows():
    print('Commentaire:', row['Commentaire'])
    print('Cluster:', row['Cluster'])
    print()
```

```python
import pandas as pd

true_labels = dataset['IsToxic']

# Créer une table croisée entre les véritables étiquettes et les étiquettes de cluster prédites
comparaison  = pd.crosstab(true_labels, labels)

# Afficher la table de comparaison
print(comparaison)
```

#  clustring avec l'algorithme DBSCAN

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import DBSCAN
import pandas as pd

# Calcul des vecteurs TF-IDF
vectorizer = TfidfVectorizer()
tfidf_vectors = vectorizer.fit_transform(col2_clean)

# Création d'un modèle de clustering DBSCAN
dbscan = DBSCAN(eps=1, min_samples=5)

# Entraînement du modèle sur les vecteurs TF-IDF
dbscan.fit(tfidf_vectors)

# Obtention des étiquettes de cluster pour chaque commentaire
labels = dbscan.labels_

# Création d'un nouveau dataframe contenant les commentaires et leur étiquette de cluster
clustered_data = pd.DataFrame({'Commentaire': col2_clean, 'Cluster': labels})

# Affichage des commentaires et leur étiquette de cluster
for _, row in clustered_data.iterrows():
    print('Commentaire:', row['Commentaire'])
    print('Cluster:', row['Cluster'])
    print()
```

```python
# Assumant 'dataset' est votre dataframe contenant les véritables étiquettes (IsToxic)
true_labels = dataset['IsToxic']

# Créer une table croisée entre les véritables étiquettes et les étiquettes de cluster prédites
cross_table = pd.crosstab(true_labels, labels)

# Afficher la table croisée pour visualiser la comparaison
print(cross_table)
```
